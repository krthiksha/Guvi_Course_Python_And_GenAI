```python

# creation of array 
import numpy as np
list1 = [1,2,3,4,4,6,78]
arr = np.array(list1)
arr
```

```python
# 1D (1 dimensional array)
arr_1d = np.array([1,2,3,5,4])
print(arr_1d)
```

```python
# 2D (2 dimensional array)
arr_2d = np.array([[1,2,3,5,4],['a','b','c','d','e']])
print(arr_2d)
```

```python
# 3D (3 dimensional array)
arr_3d = np.array([[1,2,3,5,4],['a','b','c','d','e'],[33,44,55,55,44]])
print(arr_3d)
```
